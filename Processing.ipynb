```python
import pandas as pd
import numpy as np
```

```python
from os import listdir
```

```python
#afficher les noms de mes fichiers 
noms_fichiers = listdir("speeches")
```

```python
noms_fichiers
```

```python
df = pd.DataFrame(index=range(len(noms_fichiers)), columns = ['Nom_du_fichier','contenu'])
df.head()
```

```python
for i,f_name in enumerate(noms_fichiers):
    f =open("speeches/"+f_name,"r",encoding="utf-8")
    f_content = f.readlines()
    f.close()
    
    df.at[i,'Nom_du_fichier'] = f_name
    df.at[i,'contenu'] = f_content
```

```python
df
```

```python
#liste des mois possible 
mois = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
mois
```

```python
def Separville(v):
    for m in mois:
        if m in v:
            ville = v[:v.find(m)]
    return ville 
```

```python
#Applicattions  de ma fonction
df['Ville']=df.Nom_du_fichier.apply(Separville)
```

```python
df.head()
```

```python
#Extraire la date 
def SeperateDate(c):
    dates = c.Nom_du_fichier[len(c.Ville):c.Nom_du_fichier.find(".txt")]
    return dates
```

```python
df['Date'] = df.apply(SeperateDate, axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%b%d_%Y')
df.head()
```

```python
df.drop("Nom_du_fichier",axis=1,inplace=True)
```

```python
df
```

```python
#Année
df['Année'] = df.Date.dt.year
#Mois
df['Mois'] = df.Date.dt.month
#Jour
df['Day'] = df.Date.dt.day
#Jours de la semaines
df['Jour_semaine'] = df.Date.dt.weekday
```

```python
df.head()
```

```python
df['Weekend'] = np.where((df.Jour_semaine == 5) | (df.Jour_semaine == 6),"OUI","NON")
```

```python
df
```

```python
df.to_csv("cleaned.csv",index=False)
```
